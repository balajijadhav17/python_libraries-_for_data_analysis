```python
batsman = pd.read_csv("sharma-kohli.csv")
batsman

plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

```python
# label title,xlabel,ylabel
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
plt.title("IPL run comparison between virat and rohit")
plt.xlabel("season")
plt.ylabel("runs scored")

```

```python
# color,line width and line styles ,marker
plt.plot(batsman["index"],batsman["V Kohli"],color = "blue",marker='o', linestyle='dashed',linewidth=5)
plt.title("IPL run comparison between virat and rohit")
plt.xlabel("season")
plt.ylabel("runs scored")
```

```python
# legend
plt.plot(batsman["index"],batsman["V Kohli"],color = "blue",marker='o', linestyle='dashed',linewidth=3,label="Virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color = "green",marker='o', linestyle='dashed',linewidth=3,label="Rohit")
plt.title("IPL run comparison between virat and rohit")
plt.xlabel("season")
plt.ylabel("runs scored")
plt.legend()
```

```python
# grid
plt.plot(batsman["index"],batsman["V Kohli"],color = "blue",marker='o', linestyle='dashed',linewidth=3,label="Virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color = "green",marker='o', linestyle='dashed',linewidth=3,label="Rohit")
plt.title("IPL run comparison between virat and rohit")
plt.xlabel("season")
plt.ylabel("runs scored")
plt.legend()
plt.grid()
plt.show()
```

# Scatter Plots

```python
x = np.linspace(-10,10,50)

y = 10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
plt.show()
```

```python
df = pd.read_csv("batter.csv")
df = df.head(50)
df
```

```python
plt.scatter(df["avg"],df["strike_rate"],marker="o",color="cyan")
plt.title("avg and strikrate analysis of ipl batsman")
plt.xlabel("avg")
plt.ylabel("strike_rate")
plt.grid()
plt.show()
```

# Bar chart

```python
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
df = pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"],df["2015"])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='cyan',label="2015")
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red',label="2016")
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue',label="2017")

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.legend()
plt.show()
```

```python
df.set_index('batsman').plot(kind='bar')
plt.show()
```

```python
# xticks
children = [10,20,40,10,30]
colors = ['red red red red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
plt.xticks(rotation=70)
plt.show()
```
