```python
import numpy as np
```

# np.sort

```python
a=np.array([2,3,55,1,0,6]) # 1D in ascending order
print(np.sort(a))
```

```python
a=np.array([2,3,55,1,0,6]) # 1D in descending order
print(np.sort(a)[::-1])
```

```python
a=np.random.randint(10,30,12).reshape(3,4)
print(a,"\n")
print(np.sort(a),"\n") # row wise sord
print(np.sort(a,axis=0)) # column wise sort 
```

# np.concatenate

```python
a=np.random.randint(10,30,6)
b=np.random.randint(10,30,6)
print(a)
print()
print(b)
print()
print(np.concatenate((a,b)))  # 
```

```python
a=np.random.randint(10,30,6).reshape(2,3)
b=np.random.randint(10,30,6).reshape(2,3)
print(a)
print()
print(b)
print()
print(np.concatenate((a,b)))  
print()
print(np.concatenate((a,b),axis=1))
```
