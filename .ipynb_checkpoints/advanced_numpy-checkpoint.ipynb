# Numpy vs List

```python
#list speed

a=[i for i in range(100000000)]
b=[i for i in range(100000000,200000000)]
c=[]
import time
start=time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])
time_in_second_for_list = time.time()-start
print(time_in_second_for_list)
```

```python
# Numpy speed
a=np.arange(100000000)
b=np.arange(100000000,200000000)

start=time.time()
c=a+b
time_in_second_for_numpy = time.time()-start
print(time_in_second_for_numpy)
```

```python
time_in_second_for_list/time_in_second_for_numpy
```

```python
#memory
a=[i for i in range(10000000)]
import sys
sys.getsizeof(a)  
#size in bytes
```

```python
b= np.arange(10000000)
sys.getsizeof(b)
#size in bytes
```

```python
b= np.arange(10000000,dtype=np.int32)
sys.getsizeof(b)
#size in bytes
```

# Advance Indexing

```python
# Fancy indexing
a=np.random.randint(1,100,24).reshape(6,4)
a
```

```python
a[[1,3,5]]  # access rows
```

```python
a[:,[1,2]]
```

```python
# Boolean indexing

a[a<10]
```

```python
a[a%2==0]
```

```python
a[(a >60) & (a%2==0)]
```

# Broadcasting

```python
a=np.arange(1,10).reshape(3,3)
b=np.arange(3)
print(a)
print()
print(b)
print("\n")
c=a+b

print(c)
```

```python
a=np.arange(3).reshape(1,3)
b=np.arange(3).reshape(3,1)

print(a)
print()
print(b)
print("\n")
c=a+b

print(c)
```

```python
a=np.arange(3).reshape(1,3)
b=np.arange(4).reshape(4,1)

print(a)
print()
print(b)
print("\n")
c=a+b

print(c)
```

```python
a=np.array([1])
b=np.arange(4).reshape(2,2)
print(a)
print()
print(b)
print("\n")
c=a+b

print(c)
```
