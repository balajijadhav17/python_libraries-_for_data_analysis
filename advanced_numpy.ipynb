# Numpy vs List

```python
#list speed

a=[i for i in range(10000000)]
b=[i for i in range(10000000,20000000)]
c=[]
import time
start=time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])
time_in_second_for_list = time.time()-start
print(time_in_second_for_list)
```

```python
# Numpy speed
import numpy as np
a=np.arange(100000000)
b=np.arange(100000000,200000000)

start=time.time()
c=a+b
time_in_second_for_numpy = time.time()-start
print(time_in_second_for_numpy)
```

```python
import numpy as np

data = np.array([
    ('Alice', 25, 50000),
    ('Bob', 30, 60000)
], dtype=[('name', 'U10'), ('age', 'i4'), ('salary', 'f4')])

print(data)
```

# Working with mathematical formula

```python
# sigmoid

def sigmoid(array):
    return 1/(1+np.exp(-(array)))
a=np.arange(20)
sigmoid(a)
```

```python
# mean square error
def mse(actual,predicted):
    return np.mean((actual-predicted)**2)
a=np.random.randint((1,100,20))
p=np.random.randint((1,100,20))

mse(a,p)
```

# Graph plotting

```python
import matplotlib.pyplot as plt
```

```python
#x=y
x=np.linspace(-10,10,100)
y=x
plt.plot(x,y)
plt.title("Straight line passing through origin")
plt.grid()
plt.show()
```

```python
#x=y**2
y=x**2
plt.plot(x,y,color="cyan",)
plt.title("Parabola")
plt.grid()
plt.show()
```

```python
# sin
y=np.sin(x)
plt.plot(x,y)
plt.grid()
```

```python
# xlog(x)
y=x*np.log(x)
plt.plot(x,y)
plt.grid()
```

```python
# sigmoid
y=1/(1+np.exp((-x)))
plt.plot(x,y)
plt.grid()
```
