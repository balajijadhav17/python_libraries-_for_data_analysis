# loc

```python
# loc is a label based indexing that include the end index 
```

```python
import pandas as pd

student_data = [
    ["nitish",100,90,120],
    ["ankit",80,70,100],
    ["rupesh",10,7,2],
    ["rishabh",80,50,0],
    ["amit",0,0,0],
    ["ankita",0,0,0]
]

df = pd.DataFrame(student_data, columns=["Name","IQ","Marks","Package"])
df = df.set_index("Name")
df
```

```python
# apply loc to retive rows 
df.loc["nitish"]  
# return the nitish labeled rows this is label based indexing
```

```python
df.loc["ankit":"amit"]  # amit that is end index is included 
```

```python
df.loc["ankit":"ankita":2] # gap of 2
```

# fancy indexing using loc

```python
df.loc[["nitish","rishabh","ankita"]]
```
