# read csv files

```python
courses = pd.read_csv("courses.csv")
students = pd.read_csv("students.csv")
nov = pd.read_csv("reg-month1.csv")
dec = pd.read_csv("reg-month2.csv")
matches = pd.read_csv("matches.csv")
delivery = pd.read_csv("deliveries.csv")
```

# concate()

```python
# concate vertically
pd.concat([nov,dec],ignore_index = True)  # concate the two dataframe vertically and reset the index
```

```python
# concate horizontally
pd.concat([nov,dec],axis = 1)  # concate the two dataframe horizontally
```

```python
# multiindexing
m = pd.concat([nov,dec],keys=["nov","dec"])
m
```

```python
# return just nov 
m.loc["nov"]
```
