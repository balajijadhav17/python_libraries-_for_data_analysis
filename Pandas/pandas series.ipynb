# Importing pandas library

```python
import pandas as pd
```

# Series

# create Series from list

```python
#string
country = ["india","russia","iran"]
s=pd.Series(country)
print(s)
```

```python
#number
runs=[22,55,80]
s=pd.Series(runs)
print(s)
```

# Custome indexing

```python
marks=[66,88,99]
subjects = ["hindi","marathi","math"]
s=pd.Series(marks,index=subjects)
print(s)
```

```python
marks=[66,88,99]
s=pd.Series(marks,index=["a","b","c"])
print(s)
```

```python
marks=[66,88,99]
subjects = ["hindi","marathi","math"]
s=pd.Series(marks,index=subjects,name="marks of balaji")
print(s)
```

# create series from dict

```python
price_of_product = {"pen":5,"pencil":3,"bag":300}
s=pd.Series(price_of_product)
print(s)
```

# create series with numpy

```python
import numpy as np
a=np.arange(5)
s=pd.Series(a)
print(s)
```

```python
import numpy as np
s=pd.Series(np.random.random(5))
print(s)
```

```python
import numpy as np
s=pd.Series(np.zeros(5))
print(s)
```

```python
import numpy as np
s=pd.Series(np.ones(5))
print(s)
```

```python
import numpy as np
s=pd.Series(np.linspace(1,10,4))
print(s)
```
