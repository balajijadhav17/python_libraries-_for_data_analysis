```python
import pandas as pd
import numpy as np
```

# string methods 

```python
df = pd.read_csv("titanic.csv")
df.head()
```

# lower()

```python
df["Name"].str.lower()
```

# upper()

```python
df["Sex"].str.upper()
```

# title()

```python
df["Name"].str.title()
```

# capitalize()

```python
df["Name"].str.capitalize()
```

# len()

```python
df["Name"].str.len()
```

# strip()

```python
# strip is used to remove extra spaces (or specified characters) from the beginning and end of strings.
```

```python
df["Sex"].str.strip("f")
```

# lstrip()

```python
# Removes spaces (or specified characters) from the beginning (left side)
```

```python
df["Sex"].str.lstrip("m")
```

# rstrip()

```python
# Removes spaces (or specified characters) from the end (right side)
```

```python
df["Sex"].str.rstrip("e")
```

# split()

```python
df["lastname"] = df["Name"].str.split(",").str.get(0)
df
```

# replace()

```python
df['Sex'].str.replace(r'\bmale\b', 'MALE', regex=True)
```

# startwith()

```python
df[df["Name"].str.startswith("A")]
```

# endswith()

```python
df[df["Name"].str.endswith("A")]
```

# isdigit()

```python
df[df["Sex"].str.isdigit()]
```

# isalpha()

```python
df[df["Sex"].str.isalpha()]
```

# contains()

```python
df[df["Name"].str.contains("john",case=False)]
```

```python
# Namw with start and end character is vowel
df[df["lastname"].str.contains('^[aeiouAEIOU].*[aeiouAEIOU]$', case=False, na=False)]
```
