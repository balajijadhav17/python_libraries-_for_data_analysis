# Creating timestamp object

```python
print(pd.Timestamp("2026/04/29"))
print()
#Timestamp('2026-04-29 00:00:00') (YYYY-MM-DD   HH:MM:SS)
print(type(pd.Timestamp("2026/04/29")))
print()
print(pd.Timestamp("2026-04-29"))
print()
print(pd.Timestamp(2026,4,29))
```

```python
# using text
print(pd.Timestamp("5th April 2026"))
```

```python
# providing time also
pd.Timestamp(2026,4,29,10,25,55)
```

```python
# using datetime object
import datetime as dt
dt.datetime(2026,4,29)
```

```python
x = pd.Timestamp(2026,4,29,10,30,50)
print(x)
```

```python
# extracting attributes
x = pd.Timestamp(2026,4,29,10,30,50)
print(x)
print("\n")
print("year:",x.year)
print("month:",x.month)
print("day:",x.day)
print("hour:",x.hour)
print("minute:",x.minute)
print("second:",x.second)
```

# date_range method

```python
# generate daily dates in a given range
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="D")
```

```python
# alternate days in a given range
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="3D")
```

```python
# return just business days
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="B")
```

```python
# one week per day
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="W")
```

```python
# one week per day
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="W-MON")
```

```python
# hourly data(factor)
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="h")
```

```python
# month start
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="MS")
```

```python
# month end
pd.date_range(start = "2026-04-30", end = "2026-05-30",freq="ME")
```

```python
# year start
pd.date_range(start = "2026-04-30", end = "2027-05-30",freq="YS")
```

```python
# year end
pd.date_range(start = "2026/02/01", end = "2027/05/30",freq="YS")
```

```python
# using periods
pd.date_range(start = "2026-04-30",periods=25,freq="D")
```

```python
import pandas as pd

data = {
    "Date": ["2026-01-01","2026-01-02","2026-01-03","2026-01-04","2026-01-05"],
    "Sales": [100, 150, 130, 170, 160]
}

df = pd.DataFrame(data)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.set_index("Date", inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt

df["Sales"].plot(figsize=(10,5), title="Sales Over Time")
plt.show()
```

```python
df.resample("ME").sum()
```

```python
df["Year"] = df.index.year
df["Month"] = df.index.month
df["Day"] = df.index.day
df["Day_Name"] = df.index.day_name()
```
