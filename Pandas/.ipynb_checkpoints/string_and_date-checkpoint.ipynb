```python
import pandas as pd
import numpy as np
```

# string methods 

```python
df = pd.read_csv("titanic.csv")
df.head()
```

# lower()

```python
df["Name"].str.lower()
```

# upper()

```python
df["Sex"].str.upper()
```

# title()

```python
df["Name"].str.title()
```

# capitalize()

```python
df["Name"].str.capitalize()
```

# len()

```python
df["Name"].str.len()
```

# strip()

```python
# strip is used to remove extra spaces (or specified characters) from the beginning and end of strings.
```

```python
df["Sex"].str.strip("f")
```

# lstrip()

```python
# Removes spaces (or specified characters) from the beginning (left side)
```

```python
df["Sex"].str.lstrip("m")
```

# rstrip()

```python
# Removes spaces (or specified characters) from the end (right side)
```

```python
df["Sex"].str.rstrip("e")
```
