```python
import pandas as pd 
import numpy as np
```

# Multiindex

```python
value = [("cse",2022),("cse",2021),("cse",2019),("aiml",2022),("aiml",2021),("aiml",2019)]

multiindex= pd.MultiIndex.from_tuples(value)
```

```python
s=pd.Series([1,2,3,4,5,6],index = multiindex)
s
```

```python
s["cse"]
```

```python
s[("aiml",2021)]
```

# unstack()

```python
# convert series to Dataframe
print(s.unstack())
# print(dF)
# print()
# print(type(df))
print
```
