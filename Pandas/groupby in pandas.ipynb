```python
import pandas as pd
```

```python
movies = pd.read_csv("imdb-top-1000.csv")
movies.head(1)
```

# groupby on single columns

```python
genres = movies.groupby("Genre")
```

```python
# appling builtin aggregation function on groupby object 
# genres.sum(numeric_only = True)
genres.mean(numeric_only = True)
# genres.max(numeric_only = True)
# genres.min(numeric_only = True)
# genres.std(numeric_only = True)
# genres.var(numeric_only = True)
# genres.count()

# you can apply any groupby function
```

```python
# find top 3 genres by total earning
genres.sum()[["Gross"]].sort_values("Gross",ascending = False).head(3)
```

```python
# find top 3 genres by total earning
# another way to solve this Quetion
movies.groupby("Genre")["Gross"].sum().sort_values(ascending = False).head(3)

#this is the best way
```

```python
# find the genres with highest avg IMDB_rating
movies.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending = False).head(1)

#this is the best way
```

```python
# find director with most popularity
movies.groupby("Director")["No_of_Votes"].sum().sort_values(ascending = False).head(1)
```

```python
# find number of movies done my each actor
movies.groupby("Star1")["Series_Title"].count().sort_values(ascending = False)
```
