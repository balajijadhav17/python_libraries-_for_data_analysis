# GroupBy Attributes and Methods

# len

```python
# find total number of groups 
len(movies.groupby("Genre"))
```

# size

```python
# find items in each group -- size
movies.groupby("Genre").size()
```

# first()

```python
# find the 1st row of each group
genres = movies.groupby("Genre")
genres.first()
```

# last()

```python
# find the last row of each group
genres.last()
```

# nth()

```python
# find the 7th movie of each group
genres.nth(6)
```

# get_group()

```python
# find  the all movies of specific group
genres.get_group("Horror")
```

```python
genres.get_group("Film-Noir")
```

```python
# another way of same task
movies[movies["Genre"] == "Film-Noir"]
```

# agg()

```python
genres.agg({
    "Runtime":"mean",
    "IMDB_Rating":"mean",
    "No_of_Votes":"sum",
    "Gross":"sum",
    "Metascore":"min"
})
```

# passing list in agg()

```python
movies.groupby("Genre")[["IMDB_Rating","No_of_Votes"]].agg(["min","max","mean","sum"])
```

```python
# both list and dict with agg()
genres.agg({
    "Runtime":["mean","sum"],
    "IMDB_Rating":["min","max"],
    "No_of_Votes":"sum",
    "Gross":"sum",
    "Metascore":"min"
})
```

```python
# looping on groups
for group,data in genres:
    print(data)
```

```python
# find the highest rated movie of each genre
for group,data in genres:
    print(data[data["IMDB_Rating"] == data["IMDB_Rating"].max()])
```

```python
# apply()
# find number of movies starting with A for each group

def foo(group):
    return group["Series_Title"].str.startswith("A").sum()
genres.apply(foo)
```

```python
movies.sample()
```

```python
# find the normalized IMDB rating group wise

def normal(group):
    group["norm_rating"]=((group["IMDB_Rating"] - group["IMDB_Rating"].min())/(group["IMDB_Rating"].max() - group["IMDB_Rating"].min()))
    return group

genres.apply(normal)
```

```python
# Avearge rating of each group
genres["IMDB_Rating"].mean()
```

# groupby on multiple columns

```python
duo = movies.groupby(["Director","Star1"])
duo.size()
duo.get_group(("Aamir Khan","Amole Gupte"))
```

```python
# find the most earning actor -- director combo
duo["Gross"].sum().sort_values(ascending = False).head(1)
```

```python
# find the best(in-terms of metascore(avg)) actor -- genre combo
movies.groupby(["Star1","Genre"])["Metascore"].mean().reset_index().sort_values("Metascore",ascending = False)
```

# agg on multiple groupby

```python
duo.agg({
    "IMDB_Rating": ["min", "max", "mean"]
})
```

```python
# Find average rating by Genre and Year
movies.groupby(["Genre", "Released_Year"])["IMDB_Rating"].mean()
```

```python
# Find min, max, mean rating by Genre and Year
movies.groupby(["Genre", "Released_Year"])["IMDB_Rating"].agg(["min", "max", "mean"])
```

# read csv file

```python
import pandas as pd
ipl = pd.read_csv("deliveries.csv")
ipl.head()
```

```python
ipl.shape
```

```python
ipl.columns
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending  = False).head(10)
```

```python
# find the batsman with max no of sixes
ipl[ipl["batsman_runs"] == 6][["batsman"]].value_counts().head(1)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl["batsman_runs"] == 6]

six.groupby("batsman")["batsman"].count().sort_values(ascending = False ).head(1).index[0]
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl["over"] >15]
temp_df = temp_df[(temp_df["batsman_runs"] ==4) |  (temp_df["batsman_runs"] == 6 ) ]
temp_df.groupby("batsman")["batsman"].count().sort_values(ascending = False).head(1).index[0]
```

```python
 # find V Kohli's record against all teams
temp_df = ipl[ipl["batsman"]=="V Kohli"]
temp_df.groupby("bowling_team")["batsman_runs"].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
def highest(batsman):
    temp_df = ipl[ipl["batsman"] == batsman]
    return temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending = False).head(1).values[0]
highest("AB de Villiers")
    
```
